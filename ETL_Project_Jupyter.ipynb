```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/NBA_roster_2019-2020.csv"
CurrentYearRoster_df = pd.read_csv(csv_file)
CurrentYearRoster_df.dropna(how="all")
```

```python
csv_file2 = "Resources/Seasons_Stats.csv"
AllStats_df = pd.read_csv(csv_file2)
AllStats_df
# Isolate the last year's stats
LatestStats_df = AllStats_df.loc[AllStats_df['Year'] == 2017]
```

```python
LatestStats_df.isna().sum()
NecessaryStats_df =  LatestStats_df[['Player', 'G', 'PTS', 'Tm']].copy()
NecessaryStats_df
```

```python
# LatestStats_df.nunique()
NecessaryStats_df['Duplicate?'] = NecessaryStats_df.duplicated(keep=False, subset=['Player'])
NecessaryStats_df
Joined_df = CurrentYearRoster_df.join(NecessaryStats_df.set_index('Player'), on='Name', lsuffix='_in_2020', rsuffix='_in_2017')
Joined_df
```

```python
Joined_df = Joined_df.dropna(how="all")
```

```python
Joined_df = Joined_df.rename(columns={"G": "Games in 2017", "PTS": "Points Scored in 2017"})
CurrentAnd2017_df = Joined_df.dropna(thresh=2, subset=['Games in 2017', 'Tm'])
CurrentAnd2017_df
# The below data frame is data for Players that play(ed) both in 2019-2020 and 2017
```

```python
SomeBasketballPlayers_df = CurrentAnd2017_df[['Position','Name','Rating','Raking','Age','Team','Games in 2017','Points Scored in 2017']].copy()
SomeBasketballPlayers_df
```

```python
SomeBasketballPlayers_df["2017 PPG"] = CurrentAnd2017_df["Points Scored in 2017"]/CurrentAnd2017_df["Games in 2017"]
```

```python
SomeBasketballPlayers_df = SomeBasketballPlayers_df[['Position','Name','Rating','Raking','Age','Team','2017 PPG']].copy()
SomeBasketballPlayers_df
```

```python
engine = create_engine('sqlite://', echo=False)
```

```python
SomeBasketballPlayers_df.to_sql('player', con=engine)
```
